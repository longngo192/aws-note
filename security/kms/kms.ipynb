# Aws KMS
## IS Key management service
> Using for store , management, encrypt keys that using protect user data 
## AWS KMS key
> Is logical representation for an encryption key, using for encrypt or decrypt data. Have metadata like Key ID, date create, date update, state...

## Key material
Create default by KMS, auto generate by HSMs (hard ware security module). Cant not edit, view, delete, export.
User can manualy upload. Or auto generate from custom key store.

* Private and public key is generated then give to user.
* Public key used to encrypt data out site aws.
* Private key used for decrypt data. This part is encrypt by master key

## AWS owned key
AWS owned keys are a collection of KMS keys that an AWS service owns and manages for use in multiple AWS accounts. Although AWS owned keys are not in your AWS account, an AWS service can use the associated AWS owned keys to protect the resources in your account.

You do not need to create or manage the AWS owned keys. However, you cannot view, use, track, or audit them. AWS does not charge you a monthly fee or usage fee for AWS owned keys and they do not count against the AWS KMS quotas for your account.
