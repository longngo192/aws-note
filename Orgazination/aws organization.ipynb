# Premise
Using multiple AWS account is best practice for scaling your environment.
* It provides a natural billing boundary for cost <span style="color: blue">(billing for each account) </span>
* isolates resources for security
* gives flexibility or individuals and teams
* being adaptable for new business processes

# AWS Organization
Is a collection of accounts and you can organize into hierarchy and manage centrally.

## Manager Account

* is the AWS account you use to create your organization
* From the management account, you can create, invite, remove other accounts in your organization. <span style="color: red"> is the ultimate owner of the organization </span>
* Having final control over security, infrastructure, and finance policies
* Has the role of a payer account and is responsible for paying all charges accrued by the accounts in its organization
* <span style="color: red"> You cannot change which account in your organization is the management account. </span>

## Member account

* is an AWS account, other than the management account (part of an organization)
* If you are an administrator of an organization, you can manage member accounts in the organization (create, invite, remove)
* A member account can belong to only one organization at a time.

* Is an container
* The parent container for all the accounts for your organization
* apply a policy to the root, it applies to all organizational units (OUs) and accounts in the organization.
* you can have only one root. AWS Organizations automatically creates it for you

## OU

* A container for accounts within a root
* Have another OU
* When you attach a policy to one of the nodes in the hierarchy, it flows down and affects all the branches (OUs) and leaves (accounts) beneath it.
* An OU can have exactly one parent, and currently each account can be a member of exactly one OU

## Service control policy (SCP)

* SCPs are similar to IAM permissions policies except that they don't grant any permissions
* SCPs specify the maximum permissions for an organization, organizational unit (OU), or account

## Backup policy

A type of policy that helps you standardize and implement a backup strategy for the resources across all of the accounts in your organization. In a backup policy, you can configure and deploy backup plans for your resources.

## Tag policy

In a tag policy, you can specify tagging rules for specific resources.
